## 1 Création d’une séquence discrète

```python
def Fcos(A, nu0, nue, phi, K):
    """A est l'amplitude, nu0 la frequence du cosinus, nue la frequence d'echantillonage, phi la phase initiale et K le nombre d'echantillons.
    Elle retourne k (les indices d'echantillonnage), t (les instants d'echantillonnage) et s (la sequence echantillonnée)"""
    k = np.arange(K)
    t = k*(1/nue)
    s = A*np.cos(2*np.pi*nu0*t+phi)
    return k, t, s
```

## 3 Échantillonnage et analyse fréquentielle
1. Créer une fonction permettant de calculer la TFD N -points d’une séquence temporelle, prenant en paramètres
d’entrée 

```python
def TFourier(s,nue,N):
    """s est le signal temporel, ve la frequence d'echantillonage, N le nombre de points de la TFD.
    Elle retourne n (indices), f (frequences normalisees), nu (frequences reelles), TF (TFD)"""
    n = np.arange(N)
    f = n/N
    nu = f*nue
    TF = np.fft.fft(s, N)
    return n, f, nu, TF

#Calcul de la TF
k, t, x1 = Fcos(1,800,5000,0,1000)
n1,f1,nu1, TFD1 = TFourier(x1, 5000, 1000)

#tracé de la TF
figReelle = px.line(x=n1, y=np.real(TFD1), 
                    title='partie reelle de la TF', 
                    labels={'x':'nu', 'y':'TF'}).show()
figImag = px.line(x=n1, y=np.imag(TFD1), 
                  title='partie imaginaire de la TF', 
                  labels={'x':'nu', 'y':'TF'}).show()
```
