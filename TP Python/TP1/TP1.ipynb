## 1.1 Signaux périodiques - signaux non périodiques
Tracer les signaux suivants sur l’intervalle [0, 4π] et retrouver les résultats mis en évidence à l’exercice 3 corrigé en
séance de TD.

```python
def Fx(t):
    return np.cos(3*t)+np.sin(5*t)

def Fy(t):
    return np.cos(np.sqrt(2)*t)+np.sin(3*t)

def Fz(t):
    return np.cos(t**2/2)

tmin=0
tmax=4
N=500
t=np.linspace(tmin,tmax,N)


fig = px.line(x=t, y= Fx(t), title='Signal Fx(t) fonction du temps', labels={'x':'Time', 'y':'Fx(t)'})
fig.show()
fig = px.line(x=t, y= Fy(t), title='Signal Fy(t) fonction du temps', labels={'x':'Time', 'y':'Fy(t)'})
fig.show()
fig = px.line(x=t, y= Fz(t), title='Signal Fz(t) fonction du temps', labels={'x':'Time', 'y':'Fz(t)'})
fig.show()
```

## 1.2 Inter-corrélation
On considère les signaux x(t) = cos(2πt) et y(t) = sin(2πt) pour t ∈ [0, 1].

```python
fig = px.line(x=t, y=Fx(t), title='Signal fx fonction du temps', labels={'x':'Time', 'y':'fx'})
fig.show()
fig = px.line(x=t, y=Fy(t), title='Signal fy fonction du temps', labels={'x':'Time', 'y':'fy'})
fig.show()

gamma, tau = san.correlate(Fx(t), Fy(t),1)

fig = px.line(x=tau, y=gamma, title='Inter-corrélation fx & fy', labels={'x':'temps', 'y':'Correlation'})
fig.show()
```
